```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method
```

```python
plt.plot(x,y,'r')
plt.xlabel("Xlabel")
plt.ylabel('Ylabel')
plt.title("title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("Xlabel")
axes.set_ylabel("Ylabel")
axes.set_title("title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.6,0.3])
axes1.plot(x,y)
axes2.plot(x,y)
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()  ##is used to tidy the measurements
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for co_axes in axes:
    co_axes.plot(x,y)
```

```python
axes
```

```python
 fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x,'r')
```

```python
####Figure size and DPI
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
ax = ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(6,4))
for ax in axes:
    ax.plot(x,y)
fig.savefig('save.png',dpi=200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label = 'x**2')
axes.plot(x,x**3, label = 'x **3')
ax.legend()
```

```python
### Plot Appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle='--',marker = 'o', markersize = '10', markeredgewidth = 3, markeredgecolor= 'green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
