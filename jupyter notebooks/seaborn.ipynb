```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x ='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,kde=False,split=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',strip=True)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc =tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='Blues')
```

```python
flights
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='Blues',linecolor='white',linewidth=3)
```

```python
sns.clustermap(fp,cmap='Blues',linecolor='white',linewidth='3',standard_scale=1)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',aspect=0.6,size=8)
```
